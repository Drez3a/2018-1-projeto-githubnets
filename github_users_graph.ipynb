```python
import pandas as pd
import re
import numpy as np

def create_graph_df(batches):
    return pd.concat(batches,
        ignore_index=True)


def fs_to_dataframe(fs):
#     return pd.DataFrame({'User': fs})
    return pd.DataFrame({'User': re.split('\W+', fs)})
```

```python
# load dataset

gh_users_orig = pd.read_csv("gh_users_followers.csv")
gh_users = gh_users_orig.copy()
```

```python
gh_users.head()
```

```python
gh_users.describe()
```

```python
# Normalizing
gh_users['Followers'] = gh_users['Followers'].fillna('')
gh_users['Followers'] = gh_users['Followers'].apply(fs_to_dataframe)
gh_users.head()
```

```python
gh_users['Fs'] = gh_users['Followers'].apply(pd.DataFrame({'User':gh_users['User']}).merge)
gh_users['nFs'] = gh_users['Fs'].apply(len)

# gh_users
```

```python
def create_edges_df(batches):
    return pd.concat(batches,
        ignore_index=True)

edges_df = create_edges_df(edges)
edges_df
```
