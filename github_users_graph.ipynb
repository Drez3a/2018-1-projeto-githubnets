```python
import pandas as pd
import re
import numpy as np

def create_graph_df(batches):
    return pd.concat(batches,
        ignore_index=True)


def fs_to_dataframe(fs):
#     return pd.DataFrame({'User': fs})
    return pd.DataFrame({'User': re.split('\W+', fs)})
```

```python
# load dataset

gh_users_orig = pd.read_csv("gh_users_followers.csv")
gh_users = gh_users_orig.copy()
```

```python
gh_users.head()
```

```python
gh_users.describe()
```

```python
# Normalizing
gh_users['Followers'] = gh_users['Followers'].fillna('')
gh_users['Followers'] = gh_users['Followers'].apply(fs_to_dataframe)
gh_users
```

```python
gh_users['Fs'] = gh_users['Followers'].apply(pd.DataFrame({'User':gh_users['User']}).merge)
gh_users['nFs'] = gh_users['Fs'].apply(len)

# gh_users
```

```python
gh_users
```

```python
# Teste: Usuarios com 0 seguidores: 1250
users_0_fs = [str(n) for n in gh_users.User[gh_users['nFs'] == 0]]
print(len(users_0_fs))
print(users_0_fs)
```

```python
# Ocorrencias por Quantidade de seguidores (Distribuicao do Grau)
groupby_nfs = gh_users.groupby('nFs')
groupby_nfs.agg({
  'User': ','.join,
  'Fs': lambda x: x.count(),
})
```

```python
# Number of edges nseguidores

np.sum(gh_users.nFs + 1) - gh_users.User.count()
```

```python
pairs = zip(gh_users.User, gh_users.Fs)

edges = [ pd.DataFrame([{'Source': str(f), 'Target':str(u)} for f in fs['User']]) for (u,fs) in pairs]
edges
```

```python
def create_edges_df(batches):
    return pd.concat(batches,
        ignore_index=True)

edges_df = create_edges_df(edges)
edges_df
```

```python
edges_df.to_csv('edges.csv', index=False)
```
