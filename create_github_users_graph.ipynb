```python
import pandas as pd
import re
import numpy as np

def create_users_df(batches):
    return pd.concat(
        [pd.DataFrame({'User': u}) for u in batches],
        ignore_index=True)


def fs_to_dataframe(fs):
#     return pd.DataFrame({'User': fs})
    return pd.DataFrame({'User': re.split('\W+', fs)})
```

```python
# load dataset

recife_users_1 = pd.read_csv('recife_users_with_gt_1_followers.csv')
recife_users_2 = pd.read_csv('recife_users_with_1_follower.csv')
recife_users_3 = pd.read_csv('recife_users_with_0_followers.csv')

recife_users = pd.concat(
        [recife_users_1, recife_users_2, recife_users_3],
        ignore_index=True)
gh_users = recife_users.copy()
# recife_users = recife_users[recife_users.User == 'karlafalcao']

#
gh_users.describe()
gh_users.head()
```

```python
gh_users['Followers'].dtypes
```

```python
# Filtrar seguidores

# [fs_to_dataframe(fs) for fs in gh_followers]

gh_users['Followers'] = gh_users['Followers'].apply(fs_to_dataframe)
gh_users['Followers'] = gh_users['Followers'].apply(pd.DataFrame({'User':gh_users['User']}).merge)

gh_users
```

```python
gh_users['nfs'] = gh_users['Followers'].apply(len)

# Usuarios com 0 seguidores
users_0_fs = [str(n) for n in gh_users.User[gh_users['nfs'] == 0]]
print(len(users_0_fs))
print(users_0_fs)
```

```python
groupby_nfs = gh_users.groupby('nfs')
groupby_nfs.agg({
  'User': ','.join,
  'Followers': lambda x: x.count()
})
```

```python

graph = create_users_df(edges)
graph
```
